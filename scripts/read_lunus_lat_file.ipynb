```python
fh = open("snc_exafel_2x_new_mode_mean_mirror.lat","rb")
```

```python
ba = bytearray(fh.read())
```

```python
len(ba)
```

```python
import struct
```

```python
Nx,Ny,Nz,xlo,xhi,ylo,yhi,zlo,zhi=struct.unpack('=IIIffffff',ba[0:36])
```

```python
import numpy as np
```

```python
arr=np.frombuffer(ba[36:],dtype=np.float32)
```

```python
lat=np.reshape(arr,(Nz,Ny,Nx))
```

```python
lat.shape
```

```python
im=1.*lat[80]
```

```python
im[im == -32767.0] = 0
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(im,vmin=0,vmax=2500)
```
